c.	Now apply cross-validation (cross_val_score) to do 10-fold cross-validation to estimate the out-of-training accuracy of decision tree learning for this task.

i.	What are the 10-fold cross-validation accuracy, precision, and recall?

The results in this test are different than the earlier test because in earlier test, the model was being trained and tested on the same set of data which resulted in overfitting and giving the 100% result. In 10 fold cross validation method, 10 random parts of data is chosen, out of which 9 are trained and tested on tenth. This is repeated training and testing dataset changes everytime.

ii. What are the 10 most predictive features? This can be measured by the normalized absolute difference of means for the feature between the two classes:
|μ_T-μ_F |/(σ_T+σ_F )
The larger this different, the more predictive the feature. Why do these make sense (or not)?


```python
dfT = df3[df3['Prediction'] == True]
dfF = df3[df3['Prediction'] == False]
nb = []
for i in range(len(df2.columns)):
    nb.append((abs(dfT[[i]].mean() - dfF[[i]].mean())  / dfT[[i]].std() + dfF[[i]].std()))
sorted(nb, key = itemgetter(0), reverse = True)[:10]
```

Here we see that the most predictive feature is based on the above factor involving the means of the two classes used for classification: true, false. The most predictive feature has the highest mean difference. The predictive features are divided on the basis of mean difference. The higher the mean difference the better the classification. The better classification represents that the dataset is divided in purer sets. Thus, using this factor makes sense.

iii.	What features are most predictive of a high crime rate? A low crime rate?

```python
important_features = model3.coef_
top_indices = np.argsort(important_features)[::-1]
bottom_indices = np.argsort(important_features)
print ('Top 10 Feature Ranking') 
for i in range(10):
    print ("{}. Feature {} ({})".format(i+1, col[top_indices[i]], important_features[top_indices[i]]))
print('')
print('Bottom 10 Feature Ranking')
for i in range(10):
    print ("{}. Feature {} ({})".format(i+1, col[bottom_indices[i]], important_features[bottom_indices[i]]))
```

Used coefficients of regression models for selecting and interpreting features. This is based on the idea that when all features are on the same scale, the most important features should have the highest coefficients in the model, while features uncorrelated with the output variables should have coefficient values close to zero. This approach can work well even with simple linear regression models when the data is not very noisy.

b.	Now use Ridge regression to reduce the amount of overfitting, using RidgeCV to pick the best alpha from among (10, 1, 0.1, 0.01, and 0.001).

i.	What is the estimated MSE of the model under 10-fold CV?

```python
model4 = linear_model.RidgeCV(alphas = (10, 1, 0.1, 0.01, 0.001))
model4.fit(df2, df1['ViolentCrimesPerPop'])
mse = cross_val_score(model4, df2, df1['ViolentCrimesPerPop'], scoring = 'neg_mean_squared_error' ,cv=10)
mse.mean()
```

ii.	What is the MSE on the training set (train on all the data then test on it all)?

```python
ridge_predict = model4.predict(df2)
mean_squared_error(df1['ViolentCrimesPerPop'], ridge_predict)
```

iii.	What is the best alpha?

```python
model4.alpha_
```

Alpha, also called as regularization, is used to set certain penality to reduse noise over the data. Larger the alpha value, more is the domination effect for regularization which tends the coefficients to zero. If the alpha value tends to be zero and the solution implies towards the ordinary least squares then coefficients tend to be big oscillation. Thus, it is preferable that the alpha is selected in a way that it maintains the balance between both.

Mean Squared Error (MSE) helps to determine if the model is overfitted or underfitted. The MSE value for Linear Regression here is very small thus representing very less overfitting. Thus causing an overfitting problem as each of the declared coefficient estimates will not specify the true patterns and relationship in dataset. It only presents the sample data.
Ridge regression specify prior knowledge about how the regression plays the role in estimation (determined by alpha value)
